```python
from textblob import TextBlob
from csv import reader
from csv import DictReader
from csv import reader
import matplotlib.pyplot as plt
```

```python
with open('data.csv', 'r') as read_file:
    csv_reader = reader(read_file)
    # Dictionary 
    csv_dict_reader = DictReader(read_file) 
    negative = []
    positive = []
    neutral = []
    for row in csv_reader:
        lol = row[1]
        testi = TextBlob(lol)
        # print(testi.sentiment)
        if(testi.sentiment.polarity < float(0.0)):
            negative.append(lol)
            # print(testi.sentiment)
        elif(testi.sentiment.polarity > float(0.0)):
            positive.append(lol)
        else:
            neutral.append(lol)
lenneg = len(negative)
lenpos = len(positive)
lenneu = len(neutral)
labels = 'Negative', 'Positive', 'Neutral'
sizes = [lenneg, lenpos, lenneu]
colors = ['red', 'green', 'blue']
explode = (0.1, 0, 0)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()

print("all done")
```
